```python
import pandas as pd
import numpy as np
```

```python
dados = 'csv/base/quartos.csv'
df = pd.read_csv(dados, sep=',')
df.head()
```

```python
# Remove a letra 'Q' do id_quarto e converte para inteiro
df['id_quarto'] = df['id_quarto'].str.replace('Q', '').astype(int)
df.info()
```

```python
# valor_base contém valores nulos, é preciso decidir se será feito uma média, exclusão dos dados para análise ou se manterá como 0
```

```python
# Padronização para caixa alta e correção de valores escritos errado
df['tipo_quarto'] = df['tipo_quarto'].str.upper().replace('STANDART', 'STANDARD')
df['tipo_quarto'].value_counts()
```

```python
# capacidade 0 e negativa(-1 e -2) 
#Alterar o valor negativo(-) e o 0 de acordo com a quantidade equivalente pelo tipo de quarto
#df['capacidade'] = df['capacidade'].astype(str)
#df['capacidade'] = df['capacidade'].str.replace('0', '5')
#df['capacidade'] = df['capacidade'].str.replace('-1', '5')
#df['capacidade'] = df['capacidade'].str.replace('-2', '4')
#df['capacidade'] = df['capacidade'].astype(int)
mapeamento_capacidade = {
    0: 5,
    -1: 5,
    -2: 4
}

df['capacidade'] = df['capacidade'].replace(mapeamento_capacidade)
df.groupby('capacidade').size()

```

```python
#Fazer a mediana de valor_base com o mesmo tipo de quarto
#Imputar com a mediana do mesmo tipo de quarto no mesmo hotel

df_hoteis1 = df.loc[df['id_hotel'] == 1]
df_hoteis2 = df.loc[df['id_hotel'] == 2]
df_hoteis3 = df.loc[df['id_hotel'] == 3]
df_hoteis4 = df.loc[df['id_hotel'] == 4]
df_hoteis5 = df.loc[df['id_hotel'] == 5]

df_hoteis1.groupby('tipo_quarto')['valor_base'].median().reset_index()
```

```python
#Correção valor_base nulo
#ID_HOTEL 1
df.loc[(df['id_hotel'] == 1) & (df['tipo_quarto'] == 'STANDARD') & (df['valor_base'].isnull()), 'valor_base'] = 320
#ID_HOTEL 4
df.loc[(df['id_hotel'] == 4) & (df['tipo_quarto'] == 'STANDARD') & (df['valor_base'].isnull()), 'valor_base'] = 350
#ID_HOTEL 2
df.loc[(df['id_hotel'] == 2) & (df['tipo_quarto'] == 'SUPERIOR') & (df['valor_base'].isnull()), 'valor_base'] = 420
#ID_HOTEL 5
df.loc[(df['id_hotel'] == 5) & (df['tipo_quarto'] == 'SUPERIOR') & (df['valor_base'].isnull()), 'valor_base'] = 620

```

```python
#Remover o ID hotel 9
df = df[(df['id_hotel'] <= 5)]
df['id_hotel'].value_counts()
```
