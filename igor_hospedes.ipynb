```python
import pandas as pd
import numpy as np
```

```python
dados = 'hospedes.csv'
df = pd.read_csv(dados, sep=',')
df.head()
```

```python
# Padronização do genero para M e F, convertendo todas as letras para caixa alta com upper() e substituindo as palavras necessárias para M ou F
df['genero'] = df['genero'].str.upper().replace('MASCULINO', 'M').replace('FEM', 'F')
df['genero'].value_counts()
```

```python
# Preenchimento de dados nulos por dados de 'Não informado'
df['cidade_origem'] = df['cidade_origem'].fillna('Não informado')
df['estado'] = df['estado'].fillna('NI')
df['genero'] = df['genero'].fillna('N')
df.info()
```
