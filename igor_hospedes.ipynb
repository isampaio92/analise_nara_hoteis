```python
import pandas as pd
import numpy as np
```

```python
dados = 'csv/base/hospedes.csv'
df = pd.read_csv(dados, sep=',')
df.head(10)
```

```python
df.info()
```

```python
df.describe()
```

```python
#Dados sem resposta na coluna 'cidade_origem' preenchidos como "nulo"
df['cidade_origem'] = df['cidade_origem'].fillna('Nulo')
df['cidade_origem'].value_counts()
```

```python
#Dados sem resposta na coluna 'estado' preenchidos como "nulo"
df['estado'] = df['estado'].fillna('Nulo')
df['estado'].value_counts()
```

```python
#Dados sem resposta na coluna 'genero' preenchidos como "nulo"
df['genero'] = df['genero'].fillna('Nulo')
df['genero'].value_counts()
```

```python
df['genero'] = df['genero'].str.upper().value_counts()
df['genero'] = df['genero'].replace('MASCULINO', 'M')
df['genero']


```

```python
#Exclusão de dados duplicados na base
df = df.drop_duplicates()
```

```python
# Padronização do genero para M e F, convertendo todas as letras para caixa alta com upper() e substituindo as palavras necessárias para M ou F
df['genero'] = df['genero'].str.upper().replace('MASCULINO', 'M').replace('FEM', 'F')
df['genero'].value_counts()
```

```python
#Padronização do nome com caixa inconsistente 
df['nome'] =df['nome'].str.strip().str.title()
df['nome'].head()
```

```python
#Preenchimento de emails inválidos como 'Nulo'
df['email'] = df['email'].isnull().sum() 
df['email'].value_counts()


    
```

```python
#Correção de telefones com formatos mistos 
df['telefone']

df['telefone']  
```

```python
# Preenchimento de dados nulos por dados de 'Não informado'
df['cidade_origem'] = df['cidade_origem'].fillna('Não informado')
df['estado'] = df['estado'].fillna('NI')
df['genero'] = df['genero'].fillna('N')
df.info()
```
