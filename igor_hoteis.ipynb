```python
import pandas as pd
import numpy as np
```

```python
dados = 'hoteis.csv'
df = pd.read_csv(dados, sep=',')
df
```

```python
# Nara Salvador não tem número de quartos preenchido, deve ser decidido se será melhor excluir o dado
# quartos.csv, fazer somatorio do numero de quartos de Nara Salvador
# Muda a categoria 6 estrelas para 5 estrelas
df['categoria'] = df['categoria'].str.replace('6 estrelas', '5 estrelas')
df
```

```python
# Preenche o número de quartos do Nara Salvador com o somatório do número de quartos da base quartos.csv
df['n_quartos'] = df['n_quartos'].fillna(11)
df
```

```python
# Corrige erros de digitação na coluna cidade
df['cidade'] = df['cidade'].str.replace('Rio De Janeiro', 'Rio de Janeiro').replace('Sao Paulo', 'São Paulo')
df
```
