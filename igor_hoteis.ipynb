```python
import pandas as pd
import numpy as np
```

```python
def agrupar_soma(df, qual, quant, menorMaior=False):
    return df.groupby(qual)[quant].sum().reset_index().sort_values(by=quant, ascending=menorMaior)
```

```python
def aplicar_loc(df, col_condicao, valor_condicao, sinal):
    if sinal == '==':
        return df.loc[df[col_condicao] == valor_condicao]
    elif sinal == '>':
        return df.loc[df[col_condicao] > valor_condicao]
    elif sinal == '<':
        return df.loc[df[col_condicao] < valor_condicao]
    else:
        print('Operação Inválida!')
        return
```

```python
def criar_array(df, coluna='Valor de Venda'):
    return np.array(df[coluna])
```

```python
def calcular_quartis(df, coluna='Valor de Venda'):
    array = criar_array(df, coluna)
    q1 = np.percentile(array, 25)
    q2 = np.percentile(array, 50)
    q3 = np.percentile(array, 75)

    return [q1, q2, q3]
```

```python
def calcular_distancia(df, coluna='Valor de Venda'):
    array = criar_array(df, coluna)
    media = np.mean(array)
    mediana = np.median(array)

    return (media - mediana) / mediana
```

```python
def calcular_limites(df, coluna='Valor de Venda'):
    quartis = calcular_quartis(df, coluna)

    iqr = quartis[2] - quartis[0]

    lim_sup = quartis[2] + (1.5 * iqr)
    lim_inf = quartis[0] - (1.5 * iqr)

    return [lim_sup, lim_inf]
```

```python
def mostrar_dist_lim(dist, lim):
    print(f'Distância: {dist}')
    print(f'Limite Superior: {lim[0]}')
    print(f'Limite Inferior: {lim[1]}')
```
