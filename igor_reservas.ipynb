```python
import pandas as pd
import numpy as np

# Importa as funções para os cálculos estatísticos. Ex: et.calc_iqr(df, 'coluna') - Faz o cálculo do IQR da coluna enviada
import estatistica as et
```

```python
dados = 'csv/base/reservas.csv'
df = pd.read_csv(dados, sep=',')
df.head()
```

```python
df.info()
```

```python
df['status_reserva'].value_counts()
```

```python
# Padroniza todos os valores de status_reserva para caixa alta e troca o valor CANCELADA para CANCELADO
df['status_reserva'] = df['status_reserva'].str.upper().replace('CANCELADA', 'CANCELADO')
df['status_reserva'].value_counts()
```

```python
# Remove a letra 'Q' do id_quarto e converte para inteiro
df['id_quarto'] = df['id_quarto'].str.replace('Q', '').astype(int)
df.info()
```

```python
# canal_reserva contém valores nulos
# Preenche as informações nulas com o valor 'Não informado'
df['canal_reserva'] = df['canal_reserva'].fillna('Não informado')
df.info()
```

```python
# Cria um dicionário para mapeamento dos canais de reserva e faz um replace utilizando esse mapeamento
mapeamento_canais = {
    'EXPEDIA': 'Expedia',
    'booking.com': 'Booking.com',
    'site próprio': 'Site Próprio',
    'Agencia de Viagem': 'Agência de Viagem'
}
df['canal_reserva'] = df['canal_reserva'].replace(mapeamento_canais)
df['canal_reserva'].value_counts()
```

```python
# Converte o tipo de data_checkin, data_checkout e data_reserva de string para datetime
df['data_checkin'] = pd.to_datetime(df['data_checkin'])
df['data_checkout'] = pd.to_datetime(df['data_checkout'])
df['data_reserva'] = pd.to_datetime(df['data_reserva'])
df.info()
```

```python
# Mantém apenas as linhas onde a data de checkout é maior ou igual à data de checkin
df = df[df['data_checkout'] >= df['data_checkin']]
df = df[df['data_reserva'] <= df['data_checkin']]
df.info()
```

```python
df['valor_diaria'] = df['valor_diaria'].replace(0, np.nan)
df.info()
```

```python
# Mantém apenas as linhas onde o valor da diária é positivo e não nulo
df = df[(df['valor_diaria'] >= 0)]
df.info()
```

```python
# Elimina os id_hospede inexistentes
df = df[df['id_hospede'] <= 400]
df.info()
```

```python
# Faz o cálculo dos limites superior e inferior e filtra os outliers
limites_valor_diaria = et.calc_limites(df, 'valor_diaria')
out_sup_valor_diaria = df.loc[df['valor_diaria'] > limites_valor_diaria[1]]

print(f'Limite inferior: {et.calc_limites(df, 'valor_diaria')[0]}')
print(f'Limite superior: {et.calc_limites(df, 'valor_diaria')[1]}')
```
