```python
import pandas as pd
import numpy as np
```

```python
dados = 'reservas.csv'
df = pd.read_csv(dados, sep=',')
df.head()
```

```python
df.info()
```

```python
df['status_reserva'].value_counts()
```

```python
# Padroniza todos os valores de status_reserva para caixa alta e troca o valor CANCELADA para CANCELADO
df['status_reserva'] = df['status_reserva'].str.upper().replace('CANCELADA', 'CANCELADO')
df['status_reserva'].value_counts()
```

```python
# Remove a letra 'Q' do id_quarto e converte para inteiro
df['id_quarto'] = df['id_quarto'].str.replace('Q', '').astype(int)
df.info()
```

```python
# canal_reserva contém valores nulos
# Preenche as informações nulas com o valor 'Não informado'
df['canal_reserva'] = df['canal_reserva'].fillna('Não informado')
df.info()
```

```python
# valor_diaria também contém valores nulos, é preciso decidir se será feito uma média, exclusão dos dados para análise ou se manterá como 0
# valor_diaria contém valores negativos
df['canal_reserva'].value_counts()
```
